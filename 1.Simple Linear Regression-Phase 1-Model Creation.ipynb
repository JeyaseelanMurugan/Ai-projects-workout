```python
import pandas as pd #Pandas is a library it contains the functions related to the database
```

```python
dataset=pd.read_csv("Salary_Data.csv") #Read the preprocessed data from the csv file by using the pandas function
```

```python
dataset #Display the data contained in the CSV file stored in the variable.
```

```python
#Split Input/Output 
independent=dataset[["YearsExperience"]] #Split Input data from the dataset(Refers to X)
```

```python
independent #Disply the Input data
```

```python
dependent=dataset[["Salary"]] #Split Output data from the dataset(Refers to Y)
```

```python
dependent #Display the Output data
```

```python
#Split train and test

#Import the train_test_split function directly from the model_selection module of the scikit-learn library
from sklearn.model_selection import train_test_split 

#Pass the parameter values to the train_test_split function and assign the results to four variables
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) 
```

```python
#Model creation

#Import the LinearRegression class directly from the linear_model module of the scikit-learn library
from sklearn.linear_model import LinearRegression
regressor=LinearRegression() #Create a model instance
regressor.fit(X_train,y_train) #Fit the training data
```

```python
#Inspect learned parameters 

weight=regressor.coef_   #Slope coefficients
weight #Display the value
```

```python
#Y=WX+b ---> w - weight or slope, b - bais or initial value
#Y=9360*0 + 26777
```

```python
bais=regressor.intercept_  #Inital/Minimum value Intercept
bais #Display the value
```

```python
#Evaluation Metrics
y_pred=regressor.predict(X_test) #Generate predictions (y-values) for the unseen test data.
```

```python
from sklearn.metrics import r2_score #Imports the function r2_score from sklearn metrics module.

r_score=r2_score(y_test,y_pred) #Compares the actual values (y_test) with the predicted values (y_pred).

r_score #Returns a single number --> 1.0 - perfect prediction(better performacne), 0.0 - poor performance 
```

```python
import pickle #Import pickle library to save the model
filename="finalized_model_linear.sav" #.sav for saving - Trained model will be stored
```

```python
pickle.dump(regressor, open(filename, 'wb')) #writes the trained model to the saved file.
```

```python
#The model is loaded into memory and stored as loaded_model.
loaded_model=pickle.load(open("finalized_model_linear.sav", 'rb')) #Read the binary values from the stored file

#pass in the values (X), and the model returns the predicted target (y).
result=loaded_model.predict([[0]]) #generate predictions using the trained model
result
```
